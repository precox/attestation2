Мы получили описательную статистику для каждого из штатов. Для каждого штата виден диапазон потребления электроэнергии: от минимума до максимума, а также медиана и среднее значение.

```python
#построение boxplot для обнаружения выбросов
df.boxplot(figsize=(45, 20)) #делаем визуализацию большой - ввиду большого количества регионов и, моментами, длинных их названий
plt.title('Выбросы в потреблении электроэнергии по штатам')
plt.ylabel('Потребление электроэнергии')
plt.show()
```

```python
#в качестве примера построения диаграмм по результатам анализа, предлагаю визуализировать описательную статистику
#для решения данной задачи больше всего подойдёт, по моему мнению, столбчатая диаграмма

plt.figure(figsize=(20, 8))

#рассчитаем среднее потребление по штатам
mean_consumption = df.mean()

#построим график
mean_consumption.plot(kind='bar')
plt.title('Среднее потребление электроэнергии по штатам')
plt.xlabel('Штат')
plt.ylabel('Среднее потребление электроэнергии')
plt.grid(True)
plt.show()
```

Этот график покажет средние значения потребления электроэнергии для каждого штата, что поможет выявить лидеров и аутсайдеров по потреблению.

```python
#следующий график: топ 3 по потреблению - против боттом 3 по потреблению

#получим данные для 3 штатов с самым высоким и 3 с самым низким потреблением
top_3_states = mean_consumption.head(3)
bottom_3_states = mean_consumption.tail(3)

plt.figure(figsize=(20, 8))

#построим столбчатую диаграмму для топ-3 и bottom-3 штатов
plt.bar(top_3_states.index, top_3_states.values, color='green', label='Top 3 States')
plt.bar(bottom_3_states.index, bottom_3_states.values, color='red', label='Bottom 3 States')

plt.title('Сравнение потребления электроэнергии: Топ-3 и Bottom-3 штата')
plt.xlabel('Штаты')
plt.ylabel('Среднее потребление электроэнергии')
plt.legend()
plt.grid(True)
plt.show()
```

```python
Видно явное различие между штатами с высоким уровнем потребления (например, Rajasthan) и городами с низким уровнем (например, Mizoram).
```

```python
#увеличим фигуру по вертикали для более наглядного отображения разницы между штатами с низким потреблением
plt.figure(figsize=(20, 20))
#построим столбчатую диаграмму для топ-3 и bottom-3 штатов
plt.bar(top_3_states.index, top_3_states.values, color='green', label='Top 3 States')
plt.bar(bottom_3_states.index, bottom_3_states.values, color='red', label='Bottom 3 States')

plt.title('Сравнение потребления электроэнергии: Топ-3 и Bottom-3 штата')
plt.xlabel('Штаты')
plt.ylabel('Среднее потребление электроэнергии')
plt.legend()
plt.grid(True)
plt.show()
plt.show()

#график скроллится вниз
```
